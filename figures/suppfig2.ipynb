```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
from baselines import *

import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
mpl.rcParams['ps.fonttype'] = 42

# set plot design features
font = {'family'   : 'sans-serif',
        'sans-serif' : 'Helvetica',
        'weight'   : 'regular',
        'size'     : 24}

mpl.rc('font', **font)

colors = np.array([[95,59,151],[85,180,103],[250,186,63],[150,150,150]])/255
```

#### Define x-axis labels

```python
Ns = np.array([1000, 2000, 4000, 8000, 16000, 32000, 64000, 128000, 256000])
logN = np.log10(Ns)
```

#### Estimator run times for dataset A

```python
run_times = np.load('../data/datasetA_estimator_run_times.npy')
avg_run_times = np.squeeze(np.mean(run_times,axis=1))
std_run_times = np.squeeze(np.std(run_times,axis=1))

plt.figure(figsize=(3.5,4))

plt.errorbar(logN,avg_run_times,yerr=std_run_times, color=colors[0],capsize=2,linewidth=2)
#plt.plot(logN,(Ns/1e5 + 1.6) ** 1.55 - 1.7)
plt.xlabel('$log_{10}(N)$')
plt.xticks(np.arange(3.2,5.6,1))
plt.ylabel('avg. run time (s.)')
plt.yticks(np.arange(0,8.1,2))
plt.title('dataset A', fontsize=24)
plt.savefig("saved_images/suppfig2_datasetA.pdf",transparent=True)
```

```python
run_times = np.load('../data/datasetB_estimator_run_times.npy')
avg_run_times = np.squeeze(np.mean(run_times,axis=1))
std_run_times = np.squeeze(np.std(run_times,axis=1))

plt.figure(figsize=(3.5,4))

plt.errorbar(logN,avg_run_times,yerr=std_run_times, color=colors[0],capsize=2,linewidth=2)
#plt.plot(logN,(Ns/1e5 + 1.6) ** 1.65 + 4.7)
plt.xlabel('$log_{10}(N)$')
plt.xticks(np.arange(3.2,5.6,1))
plt.ylabel('avg. run time (s.)')
plt.title('dataset B', fontsize=24)
plt.savefig("saved_images/suppfig2_datasetB.pdf",transparent=True)
#plt.yticks(np.arange(0,8.1,2))
```
